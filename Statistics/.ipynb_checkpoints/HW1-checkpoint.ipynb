```python
import scipy as sp
import pandas as pd
import numpy as np
import numpy as np
import seaborn as sns
import plotly.express as px
%matplotlib inline
import matplotlib.pyplot as plt

sns.set(
    font_scale =2,
    style      ="whitegrid",
    rc         ={'figure.figsize':(20,7)}
)

lst = list(range(0, 100))
round(np.std(lst, ddof=1), 3)
```

```python
# Первый шаг – загрузить данные и убедиться, что всё считалось правильно:
# посмотреть на столбцы, типы данных, размерность, наличие пропущенных значений.
# В каком столбце встретилось больше всего пропущенных значений?
# В качестве ответа укажите их число, после чего удалите строки с пропущенными значениями из датасета.

games = 'games.csv'
gam = pd.read_csv(games)
gam.columns = gam.columns.str.lower()
gam.isna().sum()
```

```python
gam.dtypes
```

```python
gam = gam.dropna()
gam.head(5)
```

```python
# Выведите описательные характеристики переменной Year и постройте график распределения количества выпущенных игр по годам.
sp.stats.mode(gam.year)
```

```python
gam.year.describe().to_frame()
```

```python
gam = gam.astype({'year': 'int64'})
sns.countplot(data = gam, x = 'year')
plt.xticks(rotation=45)
```

```python
# На каких платформах произошло больше всего релизов?
# Оставьте названия только тех платформ, частота встречаемости которых составляет более 7%.
# Значения укажите в порядке убывания частоты встречаемости

percentage = gam.platform.value_counts('normalize') \
    .mul(100) \
    .round() \
    .to_frame() \
    .rename(columns = { \
    'platform': 'percent'}) \
    .query('percent >= 7')
percentage.index.to_list()
```

```python
# Игры каких издателей (Publisher) встречаются в датасете чаще всего?
# В качестве ответа укажите наибольшую абсолютную частоту встречаемости.

# gam.publisher.value_counts()

sp.stats.mode(gam.publisher)
```

```python
# Сфокусируемся на играх от Nintendo.
# Посчитайте среднее, медиану и стандартное отклонение по продажам игр данного издателя в разных регионах (NA, EU, JP, Other). 
# В каком регионе медианное значение объема продаж является наибольшим?
# Hint: mean, median, std. Переменные: NA_sales и т.д. для нужных регионов.

# nintendo_stats = gam \
#    .query('publisher == "Nintendo"') \
#    .drop(columns = \
#          ['rank', \
#           'name', \
#           'platform', \
#           'year', \
#           'genre', \
#           'publisher', \
#           'global_sales']) \
#    .describe(include = 'all')

nintendo_stats = gam[['na_sales', \
                'eu_sales', \
                'jp_sales', \
                'other_sales']] \
    .agg(['mean', \
          'median', \
          'std'])

nintendo_stats
```

```python
# Теперь посмотрим на продажи Nintendo в Японии по жанрам. Постройте боксплот, где:
# по оси x будет расположен жанр игры (Genre)
# по оси у – объем продаж в Японии (JP_sales)

nintendo = gam \
    .query('publisher == "Nintendo"')

sns.boxplot(data = nintendo, x = 'genre', y = 'jp_sales')
plt.xticks(rotation=45)
```

```python
# Визуализируйте динамику изменения объема мировых продаж (Global_Sales) по годам для игр Nintendo следующих жанров:
# Fighting, Simulation, Platform, Racing, Sports.
# На каком отрезке времени игры жанра Sports были наиболее популярны?
# Иными словами, выберите наименьший отрезок, который включает в себя несколько крупных пиков.

nint_genre = nintendo \
    .query('genre in ("Fighting", \
    "Simulation", \
    "Platform", \
    "RacingSports")')

genre_sum = nint_genre.groupby(['year', 'genre'], as_index = False) \
.agg({'global_sales': 'sum'})

sns.lineplot(x ='year', y ='global_sales', hue ='genre', data = genre_sum)
```
