## ШАГ 7

Сначала построим небольшую модель всего с одним предиктором цены (price) – horsepower.

Какой процент изменчивости объясняет полученная модель? (округлите до целого)

Подсказки

- **Какая функция нам тут нужна?**
    
    `smf.ols()` из библиотеки `statsmodels` (совместно с методами `.fit()` и `.summary()`)
    
- **Где смотреть процент изменчивости?**
    
    В верхней правой части таблицы, пункт **R-squared**.
    
- **Почему мой ответ не принимается?**
    
    Убедитесь, что вы указываете ответ в процентах и округлили число!

````python
#  ЭТО ИЗ ПРОШЛОГО КОНСПЕКТА
# from scipy import stats
# import statsmodels.api as sm
# через scipy (только одномерная)
# stats.linregress(x, y)
# через statsmodels (один из вариантов)# Y = одномерный массив с ЗП, X - массив с НП
X = sm.add_constant(df_cars.price) # добавить константу, чтобы был свободный член
model = sm.OLS(df_cars.horsepower, X) # говорим модели, что у нас ЗП, а что НП
results = model.fit() # строим регрессионную прямую
print(results.summary()) # смотрим результат
````

````python
# import statsmodels.api as sm
# import statsmodels.formula.api as smf 

# способ первый
# Y = одномерный массив с ЗП, X - массив со всеми нужными нам НП

# X = sm.add_constant(X)  # добавить константу, чтобы был свободный член
# model = sm.OLS(Y, X)  # говорим модели, что у нас ЗП, а что НП
# results = model.fit()  # строим регрессионную прямую
# print(results.summary())  # смотрим результат

# способ второй, потенциально более удобный

results = smf.ols('price ~ horsepower', df_cars).fit()
print(results.summary())
````

````python
y_all = df_cars.price
x_all = df_cars.drop('price',axis =1)
x_all.head()
````

````python
# способ первый
# Y = одномерный массив с ЗП, X - массив со всеми нужными нам НП

X = sm.add_constant(x_all)  # добавить константу, чтобы был свободный член
model = sm.OLS(y_all, X)  # говорим модели, что у нас ЗП, а что НП
results = model.fit()  # строим регрессионную прямую
print(results.summary())  # смотрим результат
````

````python
y_all_2 = df_cars.price
x_all_2 = df_cars.drop('price',axis =1).drop([col for col in df_cars.columns if "company_" in col], axis=1)
x_all_2
````

````python
# способ первый
# Y = одномерный массив с ЗП, X - массив со всеми нужными нам НП

X = sm.add_constant(x_all_2)  # добавить константу, чтобы был свободный член
model = sm.OLS(y_all_2, X)  # говорим модели, что у нас ЗП, а что НП
results = model.fit()  # строим регрессионную прямую
print(results.summary())  # смотрим результат
````
