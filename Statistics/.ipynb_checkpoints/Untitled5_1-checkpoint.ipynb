```python
import pandas as pd
```

```python
cars = pd.read_csv('cars.csv')
```

```python
cars
```

```python
#Сколько пропущенных значений встретилось в датасете?
print(cars.info())
```

```python
cars['company'] = ""
```

```python
#создадим новый признак – марку автомобиля (company). Для этого используйте столбец CarName, разбейте значения ячеек по пробелу и запишите в колонку первый элемент
cars['company'] = cars['CarName'].str.split(pat = ' ', expand = True)
```

```python
#Сколько всего уникальных марок машины встречается в датасете? 
cars['company'].nunique()
```

```python
#Столбец CarName с полным названием машины удалите из датасета, а также car_ID
cars = cars.drop(columns=['CarName', 'car_ID']) 
```

```python
cars
```

```python
#Замена значений в столбце
cars['company'] = cars['company'].replace(['maxda', 'porcshce', 'toyouta', 'vokswagen', 'vw' ],['mazda', 'porsche', 'toyota', 'volkswagen','volkswagen']).str.lower() 
```

```python
cars['company'].nunique()
```

```python
#Оставить только определенные столбцы
cars = cars[['company', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength','carwidth', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower']]
```
