–≠—Ç—É –ø–æ–ª–µ–∑–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –ø—Ä–æ–±–ª–µ–º—ã:

- –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∫–æ–Ω–≤–µ—Ä—Å–∏—é (–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ø–æ–∫—É–ø–æ–∫), –≥—Ä—É–ø–ø–∏—Ä—É—è –ø–æ –≤–∞—Ä–∏–∞–Ω—Ç—É –∏ –≤–µ—Ä—Å–∏–∏ –ú–ü. –í–æ–∑–º–æ–∂–Ω–æ —É–∂–µ —Å—Ç–∞–Ω–µ—Ç —è—Å–Ω–æ –≥–¥–µ –∏–º–µ–Ω–Ω–æ –ø—Ä–∏—á–∏–Ω—ã –ø–æ–ª–æ–º–∫–∏
- –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ pvalue –¥–ª—è –∫–∞–∂–¥–æ–π –≤–µ—Ä—Å–∏–∏ –ú–ü
- –ù–∞–π–¥–∏—Ç–µ –≤–µ—Ä—Å–∏–∏, –≥–¥–µ pvalue < 0.05
 -–ò—Å–∫–ª—é—á–∏—Ç–µ –≤–µ—Ä—Å–∏—é —Å pvalue < 0.05 –∏–∑ –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
- –ü–æ–≤—Ç–æ—Ä–Ω–æ –∑–∞–ø—É—Å—Ç–∏—Ç–µ —Ä–∞—Å—á–µ—Ç FPR —á–µ—Ä–µ–∑ A/A. 
#### –¢–µ–ø–µ—Ä—å FPR < –∞–ª—å—Ñ–∞ –í—ã –Ω–∞—à–ª–∏ –ø–æ–ª–æ–º–∫—É

## –ó–∞–≥—Ä—É–∑–∏–º –¥–∞–Ω–Ω—ã–µ –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –Ω–∏—Ö

```python
path_to_file = '/mnt/HC_Volume_18315164/home-jupyter/jupyter-s-kamsarin/Statistics/hw_aa.csv'
hw_aa = pd.read_csv(path_to_file, index_col=0, sep =';')
hw_aa.head()
```

```python
hw_aa.info()
```

```python
hw_aa.groupby('experimentVariant').purchase.value_counts()
```

##### –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫ –≤ –≥—Ä—É–ø–ø–µ 1 –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ –±–æ–ª—å—à–µ —á–µ–º –≤ 0 (–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å—à–µ)

```python
group0 = hw_aa[hw_aa.experimentVariant == 0]
group0.shape
```

```python
group1 = hw_aa[hw_aa.experimentVariant == 1]
group1.shape
```

###### –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –≥—Ä—É–ø–ø—ã –ø–æ—á—Ç–∏ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ

###  –ó–∞–ø—É—Å–∫–∞–µ–º —Å–∏–º—É–ª—è—Ç–æ—Ä A/A-—Ç–µ—Å—Ç

```python
simulations = 1000
n_s = 1000
res = []

# –ó–∞–ø—É—Å–∫ —Å–∏–º—É–ª—è—Ü–∏–π A/A —Ç–µ—Å—Ç–∞
for i in tqdm(range(simulations)):
    s1 = group0['purchase'].sample(n_s, replace = False).values
    s2 = group1['purchase'].sample(n_s, replace = False).values
    res.append(stats.ttest_ind(s1, s2, equal_var = False)[1]) # —Å–æ—Ö—Ä–∞–Ω—è–µ–º pvalue

plt.hist(res, bins = 50)
plt.style.use('ggplot')
plt.xlabel('pvalues')
plt.ylabel('frequency')
plt.title("Histogram of ttest A/A simulations ")
plt.show()

# –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–æ–∂–Ω–æ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤ –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç –∞–ª—å—Ñ–∞
sum(np.array(res) <0.05) / simulations
```

–ø—Ä–∏—á–µ–º –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å—à–µ.

))

```python
hw_aa_cr = hw_aa.groupby(['version','experimentVariant'], as_index=False)\
        .agg({'uid': 'count', 'purchase': 'sum'})\
        .rename(columns={'uid' : 'uid_count','purchase':'purchase_count'}) 
hw_aa_cr['CR'] = hw_aa_cr.purchase_count/hw_aa_cr.uid_count
hw_aa_cr    
```

–í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –Ω–∞—Ä—É—à–∞–µ—Ç—Å—è —Ä–µ–ø—Ä–µ–∑–µ–Ω—Ç–∞—Ç–∏–≤–Ω–æ—Å—Ç—å —Ç–µ—Å—Ç–æ–≤—ã—Ö –≥—Ä—É–ø–ø. –í–æ–∑–º–æ–∂–Ω–æ, –ø—Ä–∏—á–∏–Ω–∞ –ø–æ–ª–æ–º–∫–∏ –∏–º–µ–Ω–Ω–æ –≤ —ç—Ç–æ–π –≤–µ—Ä—Å–∏–∏ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è.

##### –î—Ä—É–≥–∏—Ö –¥–∞–Ω–Ω—ã—Ö —É –Ω–∞–º –≤—Å–µ—Ä–∞–≤–Ω–æ –Ω–µ—Ç) –Ω–æ –Ω–µ –±—É–¥–µ–º –ø–æ–ª–æ–≥–∞—Ç—Å—è –Ω–∞ —Å–ª—É—á–∞–π, –ø—Ä–æ–≤–µ–¥–µ–º —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç—ã( –ø–æ–æ—á–µ—Ä–µ–¥–Ω–æ –¥–ª—è –∫–∞–∂–æ–π –≤–µ—Ä—Å–∏–∏ –∏ –ø–æ–æ—á–µ—Ä–µ–¥–Ω–æ –∏—Å–∫–ª—É—á–∞—è –∫–∞–∂–¥—É—é –≥—Ä—É–ø–ø—É –∏–∑ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞)

##### –ó–∞–ø—É—Å–∫–∞–µ–º —Å–∏–º—É–ª—è—Ç–æ—Ä A/A-—Ç–µ—Å—Ç –¥–ª—è –≤–µ—Ä—Å–∏–∏ v3.8.0.0

```python
simulations = 1000
n_s = 1000
res = []

# –ó–∞–ø—É—Å–∫ —Å–∏–º—É–ª—è—Ü–∏–π A/A —Ç–µ—Å—Ç–∞
for i in tqdm(range(simulations)):
    s1 = group0.query('version == "v3.8.0.0"')['purchase'].sample(n_s, replace = False).values
    s2 = group1.query('version == "v3.8.0.0"')['purchase'].sample(n_s, replace = False).values
    res.append(stats.ttest_ind(s1, s2, equal_var = False)[1]) # —Å–æ—Ö—Ä–∞–Ω—è–µ–º pvalue

plt.hist(res, bins = 50)
plt.style.use('ggplot')
plt.xlabel('pvalues')
plt.ylabel('frequency')
plt.title("Histogram of ttest A/A simulations ")
plt.show()

# –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–æ–∂–Ω–æ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤ –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç –∞–ª—å—Ñ–∞
sum(np.array(res) <0.05) / simulations
```

**–ü—Ä–æ–¥–æ–ª–∂–∞–µ–º –∏—Å–∫–∞—Ç—å –ø—Ä–∏—á–∏–Ω—É –ø–æ–ª–æ–º–∫–∏.**
